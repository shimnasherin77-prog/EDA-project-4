```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order=pd.read_csv("covid.csv")
order.head(20)
```

```python
region_cases = order.groupby('WHO Region')['Confirmed'].sum()

# Create pie chart
plt.figure(figsize=(8,8))
plt.pie(region_cases, labels=region_cases.index, autopct='%1.1f%%', startangle=90)
plt.title("Share of Confirmed COVID-19 Cases by WHO Region", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='WHO Region', data=order, palette='Set2')
```

```python
region_deaths = order.groupby('WHO Region')['Deaths'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='WHO Region', y='Deaths', data=region_deaths, palette='magma')
```

```python
top5 = order.sort_values(by='Confirmed', ascending=False).head(5)

plt.plot(top5['Country/Region'], top5['Confirmed'], marker='o', color='green')
```

```python
top5_recovered = order.sort_values(by='New recovered', ascending=False).head(5)

plt.bar(top5_recovered['Country/Region'], top5_recovered['New recovered'], color='seagreen')
```
